```python
import pandas as pd
import requests
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

### Downloading dataset from IBM Object Storage

A real estate company is planning to invest in Boston real estate. We have collected information about various areas of Boston and are tasked with created a model that can predict the median price of houses for that area so it can be used to make offers.

The dataset had information on areas/towns not individual houses, the features are

**CRIM:** Crime per capita

**ZN:** Proportion of residential land zoned for lots over 25,000 sq.ft.

**INDUS:** Proportion of non-retail business acres per town

**CHAS:** Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)

**NOX:** Nitric oxides concentration (parts per 10 million)

**RM:** Average number of rooms per dwelling

**AGE:** Proportion of owner-occupied units built prior to 1940

**DIS:** Weighted distances to Ô¨Åve Boston employment centers

**RAD:** Index of accessibility to radial highways

**TAX:** Full-value property-tax rate per $10,000

**PTRAIO:** Pupil-teacher ratio by town

**LSTAT:** Percent lower status of the population

**MEDV:** Median value of owner-occupied homes in $1000s


```python
def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, "wb") as f:
            f.write(response.content)
```

```python
dataset_path= "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/real_estate_data.csv"
download(dataset_path, 'real_estate_data.csv')
```

### Exploring the dataset

### Data Pre-Processing

Dropping rows with missing values

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

Splitting target feature from the dependant feature.

```python
X = df.drop(columns=["MEDV"])
Y = df["MEDV"]
```

```python
X.head()
```

```python
Y.head()
```

### Creating train and test dataset

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2, random_state=1)
```

### Classification Modeling

Use squared error as the criterion.

```python
# regression_tree = DecisionTreeRegressor(criterion = 'squared_error')
```

Or use the Absolute Error for the criterion.

```python
regression_tree = DecisionTreeRegressor(criterion = 'absolute_error')
```

```python
regression_tree.fit(X_train, Y_train)
```

### Evaluation

```python
regression_tree.score(X_test, Y_test)
```

```python
prediction = regression_tree.predict(X_test)

print("$",(prediction - Y_test).abs().mean()*1000)
```
